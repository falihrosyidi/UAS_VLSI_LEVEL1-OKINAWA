```python
import pandas as pd

# 1. Fungsi Konversi Q8.24 ke Decimal
def q8_24_to_decimal(hex_val):
    if pd.isna(hex_val) or not isinstance(hex_val, str):
        return None
    try:
        # Konversi string hex ke integer 32-bit
        value = int(hex_val, 16)
        # Menangani Two's Complement (Signed)
        if value & (1 << 31):
            value -= (1 << 32)
        # Membagi dengan 2^24 untuk mendapatkan nilai desimal
        return value / (2**24)
    except ValueError:
        return None

# 2. Load Dataset
# Sesuaikan nama file jika Anda menjalankannya di komputer lokal
file_path = 'Hasil_Testbench.xlsx - Sheet1.csv'
df = pd.read_csv(file_path)

# 3. Identifikasi Kolom Target (Kolom yang mengandung kata 'Hex')
hex_columns = [col for col in df.columns if 'Hex' in col]

# 4. Proses Konversi
# Kita akan membuat kolom baru dengan awalan 'Dec_' untuk setiap hasil konversi
for col in hex_columns:
    new_col_name = f'Dec_{col.replace(" ", "_")}'
    df[new_col_name] = df[col].apply(q8_24_to_decimal)

# 5. Pembersihan Tampilan (Opsional: Menghapus kolom kosong 'Unnamed')
cols_to_show = [col for col in df.columns if not col.startswith('Unnamed')]
df_result = df[cols_to_show]

# 6. Menampilkan Hasil
print("Hasil Konversi Q8.24:")
display(df_result.head(10)) # Menampilkan 10 baris pertama

# 7. Simpan Hasil ke Excel/CSV baru jika diperlukan
# df_result.to_csv('Hasil_Konversi_Decimal.csv', index=False)
```
