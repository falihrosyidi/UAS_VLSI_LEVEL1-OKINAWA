```python
def q8_24_to_decimal(hex_str):
    # 1. Konversi hex ke integer
    value = int(hex_str, 16)
    
    # 2. Tangani Signed Integer (Two's Complement)
    # Jika bit ke-31 adalah 1, maka ini bilangan negatif
    if value & (1 << 31):
        value -= (1 << 32)
    
    # 3. Bagi dengan 2^24 untuk mendapatkan nilai desimal
    decimal_value = value / (2**24)
    
    return decimal_value

# # Contoh Penggunaan:
# hex_input = "01400000" # Contoh: 1.25 dalam Q8.24
# result = q8_24_to_decimal(hex_input)

# print(f"Hex: {hex_input}")
# print(f"Decimal: {result}")

# # Contoh Angka Negatif
# hex_neg = "FE800000" # Contoh: -1.5 dalam Q8.24
# print(f"Hex Negatif: {hex_neg} -> Decimal: {q8_24_to_decimal(hex_neg)}")
```
